```python
import pandas as pd
from IPython.display import display, HTML

# Read the dataset
dataset = pd.read_csv("stellardata.csv")
dataset.head(0)
dataset.shape

# dataset.drop_duplicates(subset ="hostname",keep = False, inplace = True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Format the points
transparency = 0.6
size = 5

# Draws the scatter plot

plot = plt.scatter(dataset.st_teff, dataset.st_lum, s=size, edgecolors='none', alpha=transparency)
plt.xlim(10000,2000)
plt.title("H-R Diagram")
plt.ylabel("Luminocity")
plt.xlabel("Temperature");
```

```python
def filtered_star_assigner (row):
   if row['st_teff'] >= 5500 and row['st_teff'] <= 6300 and row['st_lum'] >= -0.5 and row['st_lum'] <= 0.5:
      return 'red'
   return 'blue'

def filtered_star_assigner_value (row, column):
    if row[column] == 'red':
        return 1
    elif row[column] == 'blue':
        return 0
    elif row[column] == 'green':
        return 2
    
dataset['filtered_color'] = dataset.apply (lambda row: filtered_star_assigner(row), axis=1)
dataset['filtered_color_value'] = dataset.apply (lambda row: filtered_star_assigner_value(row, 'filtered_color'), axis=1)
dataset = dataset.sort_values('filtered_color_value')

display_dataset = dataset[(dataset['filtered_color'] == 'red')].copy()
display_dataset.rename(columns = {'st_teff':'Temperature', 'st_lum':'Luminocity'}, inplace = True)
display_dataset.drop_duplicates(subset ="hostname",keep = False, inplace = True)
```

```python
# Draws the scatter plot of filtered stars
plot = plt.scatter(dataset.st_teff, dataset.st_lum, c = dataset.filtered_color, s=size, alpha=transparency)
plt.xlim(10000,2000)
plt.title("H-R Diagram (Filtered Stars)")
plt.ylabel("Luminocity")
plt.xlabel("Temperature");
# with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
#     display(display_dataset[['Temperature', 'Luminocity']])
```

```python
import matplotlib.ticker as ticker

filtered_dataset = dataset[dataset['st_teff'].between(2000, 10000)]

fig, ax1 = plt.subplots(1, 1)
ax2 = ax1.twinx()
ax1.scatter(filtered_dataset.st_teff, -2.5 * filtered_dataset.st_lum + 4.83, alpha=transparency/2)
ax2.scatter(filtered_dataset.st_teff, np.power(10, filtered_dataset.st_lum), alpha=transparency/2)
ax2.set_yscale('log')
plt.xlim(11000,1000)
ax1.invert_yaxis()

ax3 = ax1.twiny()
plt.xlim(11000,1000)
ax3.xaxis.set_major_formatter(ticker.NullFormatter())
ax3.xaxis.set_minor_locator(ticker.FixedLocator([2400,3700,5200,5999,7000,10001]))
ax3.xaxis.set_minor_formatter(ticker.FixedFormatter(["M0","K0","G0","F0","A0","B0"]))

plt.title("H-R Diagram")
ax1.set_ylabel("Absolute Magnitude")
ax2.set_ylabel("Luminocity")
ax3.set_xlabel("Spectral Class")
ax1.set_xlabel("Temperature(K)")
```
